```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=dataset[['charges']]
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
#from xgboost import XGBRegressor
#regressor=XGBRegressor(n_estimators=200,learning_rate=0.5,max_depth=2,subsample=0.99,colsample_bytree=0.99,random_state=0)
#regressor=regressor.fit(X_train,y_train)
```

```python
from lightgbm import LGBMRegressor
regressor=LGBMRegressor(n_estimators=200,learning_rate=0.5,max_depth=2,subsample=0.8,colsample_bytree=0.8,boosting_type="dart",random_state=0)
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
